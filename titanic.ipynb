```python
#Importando bibliotecas
import pandas as pd
```

```python
#Realizando a leitura dos arquivos
df = pd.read_csv("train.csv") #Arquivo de treinamento do modelo
```

```python
#Visualização e tratamento de dados
df.head()
```

```python
df.info()
```

```python
#Já que as colunas Age e Cabin estão com bastante valores faltantes, não agrega na nossa análise
```

```python
#Colunas como PassengerID(Identificação do passageiro), Fare(Tarifa paga), Name(nome), Ticket(Bilhete) não serão de importância
#Já que as colunas Age e Cabin estão com bastante valores faltantes, não agrega na nossa análise
df = df.drop('Age', axis=1)
df = df.drop('Cabin', axis=1)
df = df.drop('PassengerId', axis=1)
df = df.drop('Name', axis=1)
df = df.drop('Ticket', axis=1)
df = df.drop('Fare', axis=1)
```

```python
df.head()
```

```python
#Agora precisamos codificar as colunas sex e embarked para números

from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

df['Sex'] = encoder.fit_transform(df['Sex'])
df['Embarked'] = encoder.fit_transform(df['Embarked'])
```

```python
df.head()
```

```python
df.info()
```

```python
#Criação do modelo de IA
```

Para esse tipo de problema o melhor modelo a ser usado é um modelo de *CLASSIFICAÇÃO*, isso porque, o objetivo de um modelo de classificação é prever um alvo categórico, uma classe. No nosso projeto queremos prever se determinado passageiro, vai prever a classe 0 (Não sobreviveu) ou a classe 1 (Sobreviveu)

Dentro dessa área existe diversos modelos de Machine Learning para classificação. Nesse projeto iremos utilizar um modelo 
de aprendizado baseado em Árvore de Decisão.

Como funciona?

-Aprendizado de máquina baseado em Árvore de Decisão, realiza uma estrutura de segmentação de dados baseado em *"nós"*. Imagine
uma árvore, e nessa árvore em especial a raiz está no topo, e da raiz vai surgindo nós resultando em um *"terminal", ou "folha"*.
-Cada nó interno da árvore contem um teste, uma pergunta, para os atributos, e de acordo com o valor dos atributos, resulta em 
um novo terminal.

Veja o exemplo abaixo:

<img src="https://blog.somostera.com/hs-fs/hubfs/Blog_free_images/decision-tree.png?width=1007&name=decision-tree.png">

Nesse exemplo queremos prever se baseado na oferta de varios produtos, devemos comprar ou não. Começamos o modelo com o *NÓ RAIZ*
O atributo que é testado nesse nó é "Tenho dinheiro para algo além do vestido?" e baseado no valor desse atributo (SIM ou NÃO)
surgem novas conexões que resultam em novos *nós*

```python
#Preparando os dados
```

```python
#Indicando x e y
# y é a variável alvo
# x são as variáveis preditoras

y_treino = df['Survived']
x_treino = df.drop('Survived', axis=1)
```

```python
#Importando o modelo de aprendizado baseado em Árvore de Decisão
from sklearn.ensemble import RandomForestClassifier

#Criando o modelo
modelo = RandomForestClassifier()
```

```python
#Treinando o modelo
modeloarvore = modelo.fit(x_treino, y_treino)
```

```python
#Preparando os dados para teste
```

```python
#Lendo os dados
dftest = pd.read_csv("test.csv") #Database que contem os dados dos passageiros para o teste
dfy = pd.read_csv("gender_submission.csv") #Database que contem o resultado dos passageiros para o teste
```

```python
dftest.head()
```

```python
dfy.head()
```

```python
#Adicionando o resultado no dataframe de teste
dftest['Survived'] = dfy['Survived']
```

```python
dftest.head()
```

```python
#Apagando atributos
dftest = dftest.drop('Age', axis=1)
dftest = dftest.drop('Cabin', axis=1)
dftest = dftest.drop('PassengerId', axis=1)
dftest = dftest.drop('Name', axis=1)
dftest = dftest.drop('Ticket', axis=1)
dftest= dftest.drop('Fare', axis=1)
```

```python
#Agora precisamos codificar as colunas sex e embarked para números

dftest['Sex'] = encoder.fit_transform(dftest['Sex'])
dftest['Embarked'] = encoder.fit_transform(dftest['Embarked'])
```

```python
dftest.head()
```

```python
#Indicando x e y de teste
# y é a variável alvo
# x são as variáveis preditoras

y_teste = dftest['Survived']
x_teste= dftest.drop('Survived', axis=1)
```

```python
# Realizando a previsão dos passageiros testes
previsao = modeloarvore.predict(x_teste) 
```

```python
# Métrica para saber a acurácia (% de acertos)
from sklearn.metrics import accuracy_score

print(accuracy_score(y_teste, previsao) * 100)
```

Nosso modelo obteve uma precisão de 85%, o que é um resultado muito bom para esse projeto. Se desejarmos aumentar ainda mais essa acurácia, podemos fazer algumas modificações nos atributos selecionados para a criação do modelo, e realizando os testes de acurácia.
